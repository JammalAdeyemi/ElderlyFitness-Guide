```python
import cv2
import os
```

```python
video_folder = '../../../Data_dissertation/Exercise_videos/Stepup_knee_raise_videos'
output_folder = '../../../Data_dissertation/Exercise_images/Stepup_with_knee_raise'

# Get a list of all the files in the video folder
files = os.listdir(video_folder)

# Filter out the video files using their extensions
video_files = [file for file in files if file.endswith(('.mp4', '.avi'))]
```

```python
# Loop through each video file
for video_file in video_files:
    # Create the input and output paths
    input_path = os.path.join(video_folder, video_file)
    output_path = os.path.join(output_folder, video_file.split('.')[0])

    try:
        if not os.path.exists(output_path):
            os.makedirs(output_path)  # Create the output directory if it doesn't exist

    except OSError:
        print(f'Error: Creating directory {output_path}')

    # Initialize the video capture object with the input video file
    video_path = cv2.VideoCapture(input_path)
    current_frame = 0
    skip_interval = 0.5

    fps = video_path.get(cv2.CAP_PROP_FPS)
    skip_interval_frame = int(skip_interval * fps)

    while True:
        ret, frame = video_path.read()

        if ret:
            image_name = f'{video_file.split(".")[0]}_frame{current_frame}.jpg'
            output_file = os.path.join(output_path, image_name)

            print(f'Creating {output_file}')
            cv2.imwrite(output_file, frame)
            current_frame += 1
            
            for _ in range(skip_interval_frame):
                video_path.read() # Skipping frames
        else:
            break

    video_path.release()

cv2.destroyAllWindows()
```
